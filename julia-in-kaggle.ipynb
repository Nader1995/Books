# Step 6: Configure the julia Python Package

Configure the julia Python package to use the new Julia executable:

```python
import julia
julia.install(julia='/usr/local/bin/julia1.9')
```

In case you faced this error:

> PyCall is setup for non-default Julia runtime (executable) /usr/local/bin/julia1.9.
> To use this Julia runtime, PyJulia has to be initialized first by
>     from julia import Julia
>     Julia(runtime='/usr/local/bin/julia1.9')

# Step 7: Initialize PyJulia and Run Simple Julia Code

Initialize PyJulia with the specified runtime:

```python
from julia.api import Julia
jl = Julia(compiled_modules=False, runtime='/usr/local/bin/julia1.9')
```

> Run a simple Julia command to print a message:

```python
from julia import Main

# Run a simple Julia command to print a message
Main.eval("""
println("Hello from Julia!")
""")
```

> Run a basic Julia calculation:

```python
# Run a basic Julia calculation
result = Main.eval("""
a = 10
b = 20
a + b
""")
print("Result of Julia computation:", result)
```
